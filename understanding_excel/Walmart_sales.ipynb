_This script is responsible for creating several excel sheets for a particuarlar car seller_


_Monthly Sales Summary Report_


```python
# Importing libraries
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill
from openpyxl.utils import get_column_letter
```

```python
# Storing file path
FILENAME = '/home/nueltaj/Downloads/py/data_visualization/Walmart_Sales.xlsx'
# Reading file
ws = pd.read_excel(FILENAME)
ws.head()
```

```python
# Data cleaning
ws.info()
# Check for missing values
ws.isna().any()
# check for duplicated values
ws.duplicated()
```

```python
# convert the date column into pandas date format
ws['Date'] = pd.to_datetime(ws['Date'], dayfirst = True)
# create a  MOnth column with  monthly intervals
ws['Month'] = ws['Date'].dt.to_period('M')
```

```python
ws.head()
```

```python
# group the other columns by month
monthly_ws = ws.groupby('Month').agg({
    'Store': 'last',
    'Weekly_Sales': 'sum',
    'Holiday_Flag': 'sum',
    'Temperature': 'mean',
    'Fuel_Price': 'mean',
    'CPI':'mean',
    'Unemployment':'mean'

}).reset_index()
```

```python
monthly_ws=monthly_ws.rename(columns={'Weekly_Sales':'Monthly_Sales'})
```

```python
monthly_ws
```

```python
# loading the excel file
workbook = load_workbook('/home/nueltaj/Downloads/py/data_visualization/Walmart_Sales.xlsx')

# Sheet name
sheet = workbook['Sheet1']
```

```python
# Setting the font for the sheets
header_font = Font(bold=True, name = 'Gentium')
header_fill = PatternFill(start_color='FF0000', end_color='FF0000', fill_type='solid')

for cell in sheet[1]:
    cell.font = header_font
    cell.fill = header_fill

# Formatting the excel sheet
min_width = 12
for column_cells in sheet.iter_cols():
    max_length = 0
    column_letter = get_column_letter(column_cells[0].column)
    for cell in column_cells:
        try:
            if cell.value:
                cell_length = len(str(cell.value))
                if cell_length > max_length:
                    max_length = cell_length
        except None:
            pass
        adjusted_width = max(max_length + 2,min_width)
        sheet.column_dimensions[column_letter].width = adjusted_width
```

```python
# writing to an excel sheet 
workbook.save('Monthly_walmart_sales.xlsx')
```
