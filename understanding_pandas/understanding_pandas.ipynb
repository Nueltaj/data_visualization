```python
# importing libraries
import pandas as pd
```

```python
#storing the file path 
FILENAME ="/home/nueltaj/Downloads/py/data_visualization/AB_NYC_2019.csv"
#reading the dataset
df = pd.read_csv(FILENAME)
```

```python
#display the first 5 rows
df.head()
```

```python
#to get  info of a dataframe
df.info()
```

```python
#to check duplicated values in a DataFrame
df.duplicated()
```

```python
#to check if there is any duplicated values in the dataframe
#it prints out one value
df.duplicated().any()
```

```python
#if duplicated values exist
df=df.drop_duplicates(subset=None, ignore_index=False)
```

```python
#finding missing values
df.isnull()
#or
df.isna() #i prefer this guy
```

```python
#finding missing values but ouput in a columar format
df.isna().any()
```

```python
#to get the column-wise sum of the null values present in the dataset.
df.isna().sum()
```

```python
# Filling the missing values in a DataFrame  
df.fillna(value={  
    'name': 'Not Stated',  
    'host_name': 'Not Stated',  
    'last_review': 0  
}, inplace=True)  
```

```python
# Dropping columns in a DataFrame
df.drop(['reviews_per_month'],axis=1,inplace=True)
```

```python
#display column names
df.columns
```

```python
#to check if a column has unique values
df['id'].is_unique
```

```python
# Changing the index of a DataFrame
# it's advisable to check if the column we wish to be the index has unique values
# we use the .is_unique
# if True, we proceed to
df.set_index("id", inplace=True)
```

```python
df.head()
```

```python
# display the 5th record of the dataset
df.iloc[4]
```

```python
# Renaming Columns of a DataFrame
new_col ={'name':'listing_name','number_of_reviews':'reviews'}
df.rename(columns=new_col,inplace=True)
df.head()
```

```python
# check the datatype of a column
df["last_review"].dtype.type
print(df['last_review'].unique())  
```

```python
# Replacing non-date values (like "0") with None  
df['last_review'] = df['last_review'].replace({"0": None}) 
```

```python
# Converting the Data Type of Columns
# Since column is in date, we want to pandas datetime format
df['last_review']=pd.to_datetime(df['last_review'],format='%Y-%m-%d',errors='coerce')
```

```python
df['last_review'].dtype.type
```

```python
#Converting the Data Type 
df['host_id'] = df['host_id'].astype('int32')
df.info()
```
